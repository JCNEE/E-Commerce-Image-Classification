This notebook is connected directly to `src/generate_map_images.py`. Use it to inspect the expected outputs, run the export pipeline, and preview a few generated PNG assets.

```python
from __future__ import annotations
import sys
from pathlib import Path
from IPython.display import Image, display


cwd = Path.cwd().resolve()
candidate_roots = (cwd, cwd.parent)
PROJECT_ROOT = next(path for path in candidate_roots if (path / 'src' / 'generate_map_images.py').exists())
if str(PROJECT_ROOT) not in sys.path:

    sys.path.insert(0, str(PROJECT_ROOT))


from app.animal_catalog import RANCH_ANIMALS
from app.animal_location_data import get_animal_locations
from app.animal_range_map import (

    GENERATED_MAPS_PATH,
    get_animal_location_map_asset_path,
    get_district_map_asset_path,
    get_range_map_asset_path,
)
from src.generate_map_images import main

PROJECT_ROOT
GENERATED_MAPS_PATH
```

```python
sample_animal = next(animal for animal in RANCH_ANIMALS if animal.animal_id == 'lion')
sample_location = get_animal_locations(sample_animal.animal_id)[0]

sample_paths = {
    'lion range': get_range_map_asset_path('lion'),
    'lion animal': get_animal_location_map_asset_path(sample_animal.animal_id),
    'lion district': get_district_map_asset_path(sample_animal.animal_id, sample_location.location_id),
}

sample_paths
```

```python
# Run the existing generator script logic and preview every generated image.

main()



generated_files = sorted(path.relative_to(PROJECT_ROOT) for path in GENERATED_MAPS_PATH.rglob('*.png'))

print(f"Generated {len(generated_files)} map images.")



for relative_path in generated_files:

    print(relative_path)

    display(Image(filename=str(PROJECT_ROOT / relative_path)))
```
